##Estandariazación de los datos##

Los datos generalmente son recopilados de diferentes fuentes y formatos. La estandarización de datos es un término que se usa tamién para una determinada normalización de tipo de datos dónde se le resta el promedio y se divide por la desviación estandar 

#¿Qué es una estandarización?#

Es el proceso de transformación de los datos a un formato común, permitiendo así a los investigadores realizar comparaciones significativas. 

```python
df.head()
```

```python
#Transformar de mpg a K;/L a través de la operación matemática de multiplicar mpg por 0.425144
df['citykm/l']=df['city-mpg']*0.425144

#Verifica la transformación de los datos 
df.head()
```

La Normalización es el proceso de transformar los valores de varias variables a un rango similar. Las normalizaciones típicas incluyen escalar la variable para que el promedio de la variable sea 0, escalar la variable patra que la varianza sea 1  o escalar la variable para quie los valores de la variable oscilen entre 0a 1

Para mostrar como funciona la normalización escalaremos las columnas legth, width y height
para esto reemplazaremos el valor original por (valor original)(valor máximo)

```python
#remplace (valor original) por (valor original/(valor mmáximo)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```
