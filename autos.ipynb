##EstandariazaciÃ³n de los datos##

Los datos generalmente son recopilados de diferentes fuentes y formatos. La estandarizaciÃ³n de datos es un tÃ©rmino que se usa tamiÃ©n para una determinada normalizaciÃ³n de tipo de datos dÃ³nde se le resta el promedio y se divide por la desviaciÃ³n estandar 

#Â¿QuÃ© es una estandarizaciÃ³n?#

Es el proceso de transformaciÃ³n de los datos a un formato comÃºn, permitiendo asÃ­ a los investigadores realizar comparaciones significativas. 

```python
df.head()
```

```python
#Transformar de mpg a K;/L a travÃ©s de la operaciÃ³n matemÃ¡tica de multiplicar mpg por 0.425144
df['citykm/l']=df['city-mpg']*0.425144

#Verifica la transformaciÃ³n de los datos 
df.head()
```

La NormalizaciÃ³n es el proceso de transformar los valores de varias variables a un rango similar. Las normalizaciones tÃ­picas incluyen escalar la variable para que el promedio de la variable sea 0, escalar la variable patra que la varianza sea 1  o escalar la variable para quie los valores de la variable oscilen entre 0a 1

Para mostrar como funciona la normalizaciÃ³n escalaremos las columnas legth, width y height
para esto reemplazaremos el valor original por (valor original)(valor mÃ¡ximo)

```python
#remplace (valor original) por (valor original/(valor mmÃ¡ximo)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```
